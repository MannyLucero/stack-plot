# Stack Plot Example Using Random Numbers

```python
# import packages

import numpy as np
import matplotlib.pyplot as plt
```

Create some random data to plot

```python
x = range(1,21)
y1 = np.random.rand(20)
y2 = np.random.rand(20)

# print values to see what we have

print(x)
print(y1)
print(y2)
```

```python
# Setup legend for plot
plt.plot([],[], color='r', label ='y1')
plt.plot([],[], color='b', label ='y2')

# setup the plot
plt.stackplot(x,y1,y2, colors=['r','b'])
plt.title('Stack Plot Random Numbers')
plt.legend()
plt.show()
```
